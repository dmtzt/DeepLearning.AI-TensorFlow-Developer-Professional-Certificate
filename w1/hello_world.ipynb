# Hello World

Python with Tensorflow and Keras.

The traditional paradigm of expressing rules in a coding language may not always work to solve a problem. Instead, if we feed a computer with enough data that we describe (or label) as what we want it to recognize, then we can potentially train a system to solve a problem. 

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Model
Simplest possible neural network: 1 neuron

### Walkthrough:
- `Sequential`: a series of succesively connected layers
- `Dense`: a layer of connected neurons
- `input_shape`: the shape of the input value

### Loss functions and optimizers
1. The neural network has no idea of the relationship between X and Y.
2. The neural network makes a guess.
3. The neural networks uses the loss function to measure how good or bad the guess was.
4. The neural network gives the loss function data to the optimizer which figures out the next guess.
5. Each guess should be better than the one before.

### Convergence
As the guesses get better and better and accuracy approaches 100 percent, the term convergence is used.




```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Known data

```python
X = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Training
`Ã¨pochs`: the number of training loops the neural network will go through

1. Make a guess
2. Measure how good or bad the guesses with the loss function
3. Use the optimizer and the data to make another guess
4. Repeat

```python
model.fit(X, y, epochs=500)
```

## Prediction
Neural networks try to figure out answers for everything, therefore they deal with probability.

```python
print(model.predict([10.0]))
```
